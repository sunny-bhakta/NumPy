```python
arr+10
```

```python
A = np.array([[1, 2], [3, 4]])
A
```

```python
B = np.array([[5, 6], [7, 8]])
B
```

```python
A@B # equal to np.dot(A,B)
```

```python
np.dot(A, B)
```

```python
a = np.array([1, 2, 3, 4])
b = np.array([5, 6, 7, 8])
```

```python
np.vstack((a,b))
```

```python
np.hstack((a,b))
```

```python
np.column_stack((a,b))
```
